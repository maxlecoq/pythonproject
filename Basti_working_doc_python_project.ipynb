```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
!pip install pyfixest -q
import pyfixest as pf

path_to_data ='https://raw.githubusercontent.com/maxlecoq/pythonproject/refs/heads/main/Life-Expectancy-Data-Updated.csv'
df= pd.read_csv(path_to_data)
```

```python
df
```

```python
# prompt: can you sort the data byy countries

dfs = df.sort_values('Country')
dfs
```

```python
#Tryout Basti
ndf=dfs.groupby(["Country"], as_index=False).agg({"Life_expectancy": "mean", "GDP_per_capita": "mean","BMI": "mean"})
ndf
```

```python
ndf['GDP_normalized'] = ndf['GDP_per_capita'] / 1000
ndf
```

```python
reg1 = pf.feols('Life_expectancy ~ GDP_normalized+ BMI', data=ndf, )
```

```python
pf.etable(reg1)
```
